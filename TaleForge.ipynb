**TaleForge** is an NLP-based story generation system that uses GPT-4 to create unique narratives based on user inputs. The project allows users to define the genre, setting, main character, and key elements of their story. Additionally, users can modify the sentiment (Happy, Dark, Mysterious, Thrilling) to influence the tone of the generated content.



**Key Features**

**Natural Language Generation (NLG):** Uses GPT-4 to generate creative stories.

**Prompt Engineering:** User inputs help shape the story structure.

**Sentiment Control:** Adjusts tone using text transformation techniques.

**Interactive UI (Planned):** Will be converted into a web-based application.

This project demonstrates how NLP and AI can be leveraged for creative storytelling, offering users an engaging and interactive experience.

```python
import openai

# Set OpenAI API Key
OPENAI_API_KEY = "open ai key"

# Create OpenAI client
client = openai.OpenAI(api_key=OPENAI_API_KEY)

def generate_story(prompt):
    response = client.chat.completions.create(
        model="gpt-4",
        messages=[{"role": "system", "content": "You are a creative storyteller."},
                  {"role": "user", "content": prompt}],
        max_tokens=700
    )
    return response.choices[0].message.content
```

```python
def adjust_sentiment(story, sentiment):
    sentiment_changes = {
        "Happy": {"dark": "bright", "sad": "joyful", "fear": "hope"},
        "Dark": {"happy": "grim", "joy": "gloomy", "love": "loss"},
        "Mysterious": {"clear": "cryptic", "simple": "enigmatic", "known": "unknown"},
        "Thrilling": {"calm": "intense", "soft": "explosive", "slow": "fast-paced"}
    }

    if sentiment in sentiment_changes:
        for key, value in sentiment_changes[sentiment].items():
            story = story.replace(key, value)

    return story
```
